Q1. What is Flask Framework? What are the advantages of Flask Framework?

Answer--> Flask is a popular Python web framework that allows developers to quickly build web applications using Python. It is a lightweight framework that is easy to learn and use, making it a popular choice for building small to medium-sized web applications.

Advantages of using Flask framework include:

1 Simplicity: Flask is designed to be simple and easy to use. It has a small codebase, making it easy to understand and maintain. This also means that it is quick to set up and get started with.

2 Flexibility: Flask is a flexible framework that allows developers to customize and extend it to meet their specific needs.

3 Lightweight: Flask is a lightweight framework that has minimal dependencies, making it easy to deploy and run on a wide range of platforms.

4 Scalable: Flask is highly scalable and can be used to build applications of all sizes, from small prototypes to large-scale production systems. It can be easily integrated with other tools and frameworks to support complex applications.

5 Large Community: Flask has a large and active community of developers and users, which means that there is a wealth of resources and support available online. This makes it easy to find solutions to problems and get help when needed.

Q3. What is App routing in Flask? Why do we use app routes?

Answer--> In Flask, app routing refers to the process of mapping URLs to view functions that generate the content to be displayed on that URL. Essentially, app routing allows you to define which function should be executed when a user requests a specific URL.

We use app routes in Flask for several reasons:

1 It helps us organize our code: By using app routes, we can split up our application logic into multiple functions, each responsible for handling a specific request URL.

2 It allows us to create dynamic URLs: By using app routes with URL parameters, we can create URLs that are customized based on user input. 

3 It makes our code more modular: By separating our code into smaller functions that are responsible for specific tasks, we can make our code more modular and easier to maintain.
